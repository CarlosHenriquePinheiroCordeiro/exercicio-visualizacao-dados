```python
#1  a) Elabore Diagramas de Barras plotando a relação Ano X Matrículas e Ano X Evasão.
import matplotlib.pyplot as plt
import seaborn as sb
def graficoBarras(x, y, xlabel, ylabel, titulo):
    plt.bar(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title (titulo)
    return plt

ano        = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
matriculas = [3400, 3800, 3550, 3700, 3780, 3600, 3200, 3340, 3100, 2870, 2200, 1960]
evasao     = [230 , 175 , 110 , 350 , 315 , 327 , 280 , 335 , 295 , 390 , 450 , 184]

#graficoBarras(ano, matriculas, 'Ano', 'Matrículas', 'Ano x Matrículas').show()
#graficoBarras(ano, evasao    , 'Ano', 'Evasão'    , 'Ano x Evasão').show()

#   b) Considerando o percentual de Evasão Escolar em relação às matriculas em cada ano,
#      faça a plotagem de um gráfico de dispersão e discuta sobre a tendência dos dados.
def percentualEvasao(matriculas, evasao):
    percentual = []
    for i in range(len(matriculas)):
        percentual.append((evasao[i] * 100) / matriculas[i])
    return percentual

def montaDadosGraficoDispersao(x, y):
    dados = []
    for i in range(len(x)):
        dados.append([x[i], y[i]])
    return dados    

def graficoDispersao(arrayDados, xDados, yDados):
    return sb.scatterplot(data=arrayDados, x=xDados, y=yDados)

dadosPercentualEvasao = percentualEvasao(matriculas, evasao)
graficoDispersao(arrayDados=montaDadosGraficoDispersao(ano, dadosPercentualEvasao), xDados=ano, yDados=dadosPercentualEvasao)
```
