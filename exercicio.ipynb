```python
#2  A tabela registra as marcas das temperaturas, mínima e máxima de determinada região no
#   período de uma semana. Elabore a plotagem BoxPlot de ambas as colunas registradas e identifique os elementos que
#   constituem os indicadores das caixas.
import matplotlib.pyplot as plt
import statistics
import numpy as np

#Monta um gráfico de boxplot junto de suas estatísticas
def graficoBoxPlot(arrayDados, legenda):
    print('\nBoxplot', legenda)
    plt.boxplot(arrayDados)
    plt.title(legenda)
    plt.show()
    estatisticasBoxPlot(arrayDados)

#Mostra as estatísticas de um boxplot
def estatisticasBoxPlot(arrayDados):
    q1 = np.percentile(arrayDados, 25)
    print('1º Quartil = ', q1)
    print('Mediana    = ', statistics.median(arrayDados))
    q3 = np.percentile(arrayDados, 75)
    print('3º Quartil = ', q3)
    print('Mínimo     = ', q1-1.5*(q3 - q1))
    print('Máximo     = ', q1+1.5*(q3 - q1))
    print('\n')

minima = [12, 15, 11, 13, 14, 10, 12]
maxima = [26, 28, 30, 25, 27, 24, 29]
graficoBoxPlot(minima, 'Temperaturas Mínimas')
graficoBoxPlot(maxima, 'Temperaturas Máximas')

```

```python
#3  Considere o dataset “pessoas.csv” disponibilizado. Elabore dois diagramas com plotagens
#   diferentes para discutir a relação pessoas.peso x pessoas.altura
import pandas as pd
import matplotlib.pyplot as plt

arquivo = pd.read_csv("pessoas.csv", encoding="UTF-8", sep=",")

#Monta um gráfico de linhas
def graficoLinha(x, y, xLegenda, yLegenda, titulo):
    plt.figure(figsize=(11, 6))
    plt.style.use('ggplot')
    plt.plot(x, y)
    plt.xlabel(xLegenda)
    plt.ylabel(yLegenda)
    plt.title(titulo)
    plt.show()

#Monta um gráfico de pontos
def graficoPonto(x, y, xLegenda, yLegenda, titulo):
    plt.figure(figsize=(11, 6))
    plt.style.use('ggplot')
    plt.plot(x, y, "^")
    plt.xlabel(xLegenda)
    plt.ylabel(yLegenda)
    plt.title(titulo)
    plt.show()

x = arquivo['nome']
print("Gráficos Pessoas x Peso:")
y = arquivo['peso']
graficoLinha(x, y, "Pessoa", "Peso", "Pessoas x Peso")
graficoPonto(x, y, "Pessoa", "Peso", "Pessoas x Peso")

print("Gráficos Pessoas x Altura:")
y = arquivo['altura']
graficoLinha(x, y, "Pessoa", "Altura", "Pessoas x Altura")
graficoPonto(x, y, "Pessoa", "Altura", "Pessoas x Altura")
```

```python
#5  O arquivo “anotações.txt” contém a coleta das medidas de circunferência de várias árvores
#   destinadas a produção de madeira para indústria moveleira. Essas medidas são coletadas na
#   altura de aproximadamente 1,5m (referência utilizada para cálculo do DAP = diâmetro na
#   altura do peito). Você precisa fazer uma análise que melhor descreva estatisticamente o
#   conjunto das anotações de medidas. Para tal é necessário implementar códigos que orientem:
#       a. Qual a melhor medida padrão para uma decisão de venda das árvores?
#       b. Identificar mudanças significativas se forem constituídos grupos distintos como
#       medidas de circunferência padrão em relação ao tratamento como um único grupo?
#       c. Cuidar com anotações supostamente erradas que podem deturpar a variabilidade das
#       medidas
#       Elabore:
#           d. Diagrama de dispersão
#           e. Histogramas que permitam a visualização das classes no entorno dos quartis dos
#           conjuntos

###########################################################################################################################################
#É IMPORTANTE QUE SEJAM EXECUTADOS ESTES BLOCOS PARA QUE OS SEGUINTES RECONHEÇAM AS FUNÇÕES PRESENTES NESTE
###########################################################################################################################################
#Primeiro passo: c. Cuidar com anotações supostamente erradas que podem deturpar a variabilidade das medidas

#Abertura do arquivo
import pandas as pd
arquivo = pd.read_csv('anotacoes.txt', encoding='UTF-8', sep='\t')

#Funções para tratamento de dados: Vale constar que todo tratamento retorna um vetor com os dados tratados, sem alterar o arquivo aberto
#Retorna os dados tratados por grupo
def pegaDadosTratadosPorGrupo(arquivo):
    dados = []
    for i in arquivo:
        dados.append(trataDados(arquivo[i]))
    return dados

#Retorna um vetor tratado com piso de 80 e teto de 190
def trataDados(arrayDados):
    lista = []
    for i in arrayDados:
        if (i >= 80 and i <= 190):
            lista.append(i)
    return lista

#Junta os dados separados por grupo em apenas um vetor de dados
def juntaDados(arrayDados):
    lista = []
    for grupo in arrayDados:
        for i in grupo:
            lista.append(i)
    return lista

#Retorna, desde a abertura do arquivo, todos os dados juntos em um vetor
def pegaDadosJuntos():
    return juntaDados(pegaDadosTratadosPorGrupo(arquivo))
```

```python
#Segundo passo: b. Identificar mudanças significativas se forem constituídos grupos distintos como medidas de circunferência padrão em relação ao tratamento como um único grupo?  
import pandas as pd
import numpy as np
import statistics as st

#Média aritmética
def media(dados):
    return st.mean(dados)

#Retorna a média harmônica
def mediaHarmonica(dados):
    return st.harmonic_mean(dados)

#Retorna a variância
def variancia(dados):
    return np.var(dados)

#Retorna o desvio padrão
def desvioPadrao(dados):
    return np.std(dados)

#Monta uma tabela
def montaTabelaDados(dados, indices, colunas):
    return pd.DataFrame(data=dados, index=indices, columns=colunas)

#Retorna a Média, Média Harmônica, Variância e Desvio Padrão dos dados enviados como parâmetro
def pegaResultados(dados):
    print("Média"          , media(dados))
    print("Média Harmônica", mediaHarmonica(dados))
    print("Variância:"     , variancia(dados))
    print("Desvio Padrão:" , desvioPadrao(dados))

#Retorna uma linha para o gráfico
def montaLinhaGrafico(dados, indice):
    return [indice, media(dados), mediaHarmonica(dados), variancia(dados), desvioPadrao(dados)]

#A partir daqui, os dados são tratados para a demonstração de resultados a partir de aplicações estatísticas sobre estes, visando a resposta desta alternativa
dados = pegaDadosTratadosPorGrupo(arquivo)
print("Análise dos dados e mudanças quanto aos modos de análise, sendo estes separando os dados em grupo ou analisando todos de uma vez:")
print("Primeiramente mostrando os grupos:")
contador         = 0;
arrayDadosLinhas = []
colunas          = ["Grupo", "Média", "Média Harmônica", "Variância", "Desvio Padrão"]
linhas           = list(range(len(dados)))
for grupo in dados:
    contador = contador + 1
    print("\nGrupo", contador)
    print(grupo)
    arrayDadosLinhas.append(montaLinhaGrafico(dados=grupo, indice=contador))

print("Sendo assim, segue tabela com análises, em grupos e depois com todos os dados juntos:")
arrayDadosLinhas.append(montaLinhaGrafico(dados=juntaDados(dados), indice="Todos os dados juntos"))
linhas.append(len(linhas))
montaTabelaDados(dados=arrayDadosLinhas, indices=linhas, colunas=colunas)
```

```python
#Terceiro passo: a. Qual a melhor medida padrão para uma decisão de venda das árvores?
print("É preferível que se adote uma medida padrão considerando a análise de todos os dados juntos, não os separando por grupos, \nvisto as diferenças entre as Variâncias e os Desvios Padrões observados.")
print("Sendo assim, observando as médias harmônicas e aritméticas, é preferível que se utilize como medida a Média Harmônica, sobre todo o conjunto de dados.")
print("Tal medida foi tomada considerando que: Como temos valores distantes e se tratando de vendas, nem a mediana, e muito menos a moda, seriam ideiais como base para vendas.")
print("E, tendo que a média harmônica visa \"balancear\" a média aritmética, diante de um cenário com grandes variedades de medidas (isso que os dados já foram tratados), \nfoi tomada a Média Harmônica como medida.")
```
