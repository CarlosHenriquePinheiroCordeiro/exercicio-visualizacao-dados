```python
#2  A tabela registra as marcas das temperaturas, mínima e máxima de determinada região no
#   período de uma semana. Elabore a plotagem BoxPlot de ambas as colunas registradas e identifique os elementos que
#   constituem os indicadores das caixas.
import matplotlib.pyplot as plt
import statistics
import numpy as np

def graficoBoxPlot(arrayDados, legenda):
    print('Boxplot ', legenda)
    plt.boxplot(arrayDados)
    plt.title(legenda)
    plt.show()
    estatisticasBoxPlot(arrayDados)

def estatisticasBoxPlot(arrayDados):
    q1 = np.percentile(arrayDados, 25)
    print('1º Quartil = ', q1)
    print('Mediana    = ', statistics.median(arrayDados))
    q3 = np.percentile(arrayDados, 75)
    print('3º Quartil = ', q3)
    print('Mínimo     = ', q1-1.5*(q3 - q1))
    print('Máximo     = ', q1+1.5*(q3 - q1))
    print('\n')

minima = [12, 15, 11, 13, 14, 10, 12]
maxima = [26, 28, 30, 25, 27, 24, 29]
graficoBoxPlot(minima, 'Temperaturas Mínimas')
graficoBoxPlot(maxima, 'Temperaturas Máximas')

```

```python
#3  Considere o dataset “pessoas.csv” disponibilizado. Elabore dois diagramas com plotagens
#   diferentes para discutir a relação pessoas.peso x pessoas.altura
import pandas as pd
import matplotlib.pyplot as plt

arquivo = pd.read_csv("pessoas.csv", encoding="UTF-8", sep=",")

def graficoLinha(x, y, xLegenda, yLegenda, titulo):
    plt.figure(figsize=(11, 6))
    plt.style.use('ggplot')
    plt.plot(x, y)
    plt.xlabel(xLegenda)
    plt.ylabel(yLegenda)
    plt.title(titulo)
    plt.show()

def graficoPonto(x, y, xLegenda, yLegenda, titulo):
    plt.figure(figsize=(11, 6))
    plt.style.use('ggplot')
    plt.plot(x, y, "^")
    plt.xlabel(xLegenda)
    plt.ylabel(yLegenda)
    plt.title(titulo)
    plt.show()

x = arquivo['nome']
print("Gráficos Pessoas x Peso:")
y = arquivo['peso']
graficoLinha(x, y, "Pessoa", "Peso", "Pessoas x Peso")
graficoPonto(x, y, "Pessoa", "Peso", "Pessoas x Peso")

print("Gráficos Pessoas x Altura:")
y = arquivo['altura']
graficoLinha(x, y, "Pessoa", "Altura", "Pessoas x Altura")
graficoPonto(x, y, "Pessoa", "Altura", "Pessoas x Altura")
```
